From any website if I want to pull data using python, I need to install and import two libraries, 
1. requests
2. BeautifulSoup

Along with this, "lxml" engine is also required to convert text into HTML+CSS code.

In web scraping, we need to take request from website first, so for this assignment I have considered the website like wikipedia and othe free websites.

```python
# Import libraries

import requests

import bs4
```

## The task is to retrive the table of content data and the images from the  https://en.wikipedia.org/wiki/Tableau_Software  website page

```python
# Make a request using request library

res=requests.get("https://en.wikipedia.org/wiki/Tableau_Software")
```

```python
# The webpage content is converted into a string using request library 
res.text
```

```python
# To change the string into HTML+CSS code, we are going to use BeautifulSoup library

mytoc=bs4.BeautifulSoup(res.text,"lxml")
```

```python
# The string in request is converted into code having different class, id's of HTML and CSS
mytoc
```

After insepcting the webpage, I have found that for retriving the table of content data, I need to use "toctext" class mention in the inspect window  

```python
mytoc.select(".toctext")

# It gives the list of all the table of content
```

```python
first_topic=mytoc.select(".toctext")[0]

first_topic
```

```python
# To retrive data from the above code

first_topic.text
```

```python
# Second method

first_topic=mytoc.select(".toctext")[0].getText()

first_topic
```

```python
# To print all the text from mytoc, I need to use loop

for item in mytoc.select(".toctext"):
    
    print(item.text)
```

# To retrive the image on the webpage https://en.wikipedia.org/wiki/Tableau_Software

```python
# Make a request
res=requests.get("https://en.wikipedia.org/wiki/Tableau_Software")

# Use bs4 
myimg=bs4.BeautifulSoup(res.text,'lxml')

# Select class as "thumbimage" (class can be found by inspecting the webpage)
Im1=myimg.select('.thumbimage')

# Find the image url from the webpage
Im1[0]['src']
```

## The task is to retrive all the unique authors and tags from all pages on the website http://quotes.toscrape.com/

```python
all_authors=set()

base_url="http://quotes.toscrape.com/page/{}/"

for page in range(1,11):
    
    scrape_url=base_url.format(page)
    
    res=requests.get(scrape_url)
    
    soup=bs4.BeautifulSoup(res.text,"lxml")
    
    
    for author in soup.select(".author"):
        all_authors.add(author.text)
```

```python
all_authors
```

```python
all_tags=set()

base_url="http://quotes.toscrape.com/page/{}/"

for page in range(1,11):
    
    scrape_url=base_url.format(page)
    
    res=requests.get(scrape_url)
    
    soup=bs4.BeautifulSoup(res.text,"lxml")
    
    
    for item in soup.select(".tag"):
        all_tags.add(item.text)
```
