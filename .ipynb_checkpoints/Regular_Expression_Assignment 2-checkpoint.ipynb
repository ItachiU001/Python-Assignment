# Regular Expression using Python

The task is to find the code and glyph data on the webpage https://en.wikipedia.org/wiki/List_of_Unicode_characters. For the task, I have used web scraping and regular expression concepts to find the code and glyph. The used libraries are 
1. requests
2. bs4 (used for BeautifulSoup)
3. re (used for regular expression)


```python
# Import all the required library

import requests
import bs4
import re
```

```python
# Make a request

res=requests.get("https://en.wikipedia.org/wiki/List_of_Unicode_characters")
```

```python
# Get the whole string 

Text=res.text
```

```python
# Formulate the pattern for code

pattern_code=r'U\W\w{1,}'
```

```python
code=re.findall(pattern_code,Text)
```

```python
#code
```

```python
# Find all the code on the webpage

for code in re.findall(pattern_code,Text):
    print(code)
```

```python
# Pattern used for glyph

pattern_glyph=r'[\W]'
```

```python
# Find all the glyph on the webpage

for glyph in re.findall(pattern_glyph,Text):
    print(glyph)
```
